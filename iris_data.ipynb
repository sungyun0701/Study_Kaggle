```
import pandas as pd#pandas 임포트 iris파일을 읽어 데이터프레임으로 변환
df = pd.read_csv('iris.csv')
df.head(3) #끝은 tail
species = df['종류'].unique() #몇 종류가 있는지 알아볼때 unique씀

df['종류'].value_counts() #통계 낼때

# df.tail(3) #NaN은 데이터가 없거나 숫자가 아니다? 결손값이라고 부름
# df.isnull()

df.isnull().any(axis=0) #any는 모두다라는 뜻? // axis가 0이면 수직방향으로 확인 1이면 수평 방향으로 확인
# df.sum()
# df.isnull().sum() #결손값 합계 구하기df.isnull().any(axis=0) #any는 모두다라는 뜻? // axis가 0이면 수직방향으로 확인 1이면 수평 방향으로 확인
# df.sum()
df.isnull().sum() #결손값 합계 구하기
```

```
df2 = df.dropna(how='any', axis=0) #Nan값을 삭제 하겠다. inplace는 동기화
df2
# df.isnull().any(axis=0)
```

```
#데이터 보정
df.tail(3)
df.mean() #mean()은 평균값 median는 가운데값 std()표준 편차 , min, max
df['꽃잎 길이'] = df['꽃잎 길이'].fillna(0.480267) #값이 결손값일 결우 0으로 채우겠다
df
```

```
colmean = df.mean()
colmean
df2 = df.fillna(colmean)
df2.isnull().any(axis=0)
df2.tail(3)
```

```
xcol = ['꽃받침 길이','꽃받침 폭','꽃잎 길이','꽃잎 폭']
x = df2[xcol]
t = df2['종류']


from sklearn import tree #ai쪽 라이브러리
model = tree.DecisionTreeClassifier(max_depth = 2, random_state=0) #random_stata는 random sid라고 생각하면됨

model.fit(x,t) #모델 학습
model.score(x, t) # 학습이 끝난 모델의 정답율 계산
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, t,  # 훈련 데이터와 테스트 데이터를 구분한다
    test_size = 0.3, random_state = 0)

print(x_train.shape)  # x_train의 행수와 열수를 표시
print(x_test.shape)   # x_test의 행수와 열수를 표시

model.fit(x_train, y_train)  # 훈련 데이터로 재학습

model.score(x_test, y_test)  # 테스트 데이터의 예측 결과와 실제 답이 일치하는지의 정답율을 계산
```

```
import pickle
with open('irismodel.pkl', 'wb') as f:  # 모델을 보존한다
    pickle.dump(model, f)
```
